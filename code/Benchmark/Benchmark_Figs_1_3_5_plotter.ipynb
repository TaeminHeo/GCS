```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 12})
```

```python
data = pd.read_csv('../../data/toyproblem.csv',header=None)
```

```python
f, ax = plt.subplots(2,1,figsize=[10,4])
ax[0].plot(data.iloc[:,0],'k',linewidth=0.5)
ax[0].set_xlim(ax[0].get_xlim())
ax[0].set_ylim(ax[0].get_ylim())
var1_axis = ax[0].get_ylim()
ax[0].plot([ax[0].get_xlim()[0],0],[5,5],color='grey',lw=0.5,alpha=0.5)
ax[0].plot([ax[0].get_xlim()[0],300],[10,10],color='grey',lw=0.5,alpha=0.5)
ax[0].plot([ax[0].get_xlim()[0],600],[15,15],color='grey',lw=0.5,alpha=0.5)
ax[0].set_ylabel(r'$x_1$')
ax[0].set_yticks([5,10,15,20])

ax[1].plot(data.iloc[:,1],'k',linewidth=0.5)
ax[1].set_xlim(ax[1].get_xlim())
ax[1].set_ylim(ax[1].get_ylim())
var2_axis = ax[1].get_ylim()
ax[1].plot([ax[1].get_xlim()[0],0],[2,2],color='grey',lw=0.5,alpha=0.5)
ax[1].plot([ax[1].get_xlim()[0],300],[3,3],color='grey',lw=0.5,alpha=0.5)
ax[1].plot([ax[1].get_xlim()[0],600],[4,4],color='grey',lw=0.5,alpha=0.5)
ax[1].set_xlabel('data index')
ax[1].set_ylabel(r'$x_2$')
ax[1].set_yticks([1,2,3,4,5])

ax[0].fill(np.concatenate([[0,300], [300,0]]),
                 np.concatenate([[ax[0].get_ylim()[0],ax[0].get_ylim()[0]],[ax[0].get_ylim()[1],ax[0].get_ylim()[1]]]),
                 alpha=.25, fc='grey', ec='None')
ax[0].fill(np.concatenate([[600,1000], [1000,600]]),
                 np.concatenate([[ax[0].get_ylim()[0],ax[0].get_ylim()[0]],[ax[0].get_ylim()[1],ax[0].get_ylim()[1]]]),
                 alpha=.25, fc='grey', ec='None')
ax[1].fill(np.concatenate([[300,600], [600,300]]),
                 np.concatenate([[ax[1].get_ylim()[0],ax[1].get_ylim()[0]],[ax[1].get_ylim()[1],ax[1].get_ylim()[1]]]),
                 alpha=.25, fc='grey', ec='None')
plt.tight_layout()
#plt.savefig('../plots/bc.png',dpi=300)
plt.show()
```

```python
f, ax = plt.subplots(2,2,figsize=[10,4])
ax[0,0].plot(range(500),data.iloc[:500,0],'k',linewidth=0.5)
ax[0,0].set_ylabel(r'$x_1$')
ax[0,0].set_ylim(var1_axis)
ax[0,0].axes.xaxis.set_ticklabels([])
ax[0,0].set_yticks([3,6,9,12])

ax[0,1].plot(range(500,1000),data.iloc[500:,0],'k',linewidth=0.5)
ax[0,1].set_ylim(var1_axis)
ax[0,1].axes.yaxis.set_ticklabels([])
ax[0,1].axes.xaxis.set_ticklabels([])

ax[1,0].plot(range(500),data.iloc[:500,1],'k',linewidth=0.5)
ax[1,0].set_xlabel(r'$X_1$')
ax[1,0].set_ylabel(r'$x_2$')
ax[1,0].set_ylim(var2_axis)
ax[1,0].set_yticks([1,2,3,4,5])

ax[1,1].plot(range(500,1000),data.iloc[500:,1],'k',linewidth=0.5)
ax[1,1].set_xlabel(r'$X_2$')
ax[1,1].set_ylim(var2_axis)
ax[1,1].axes.yaxis.set_ticklabels([])

plt.tight_layout()
#plt.savefig('../plots/bc_k.png',dpi=300)
plt.show()
```
