```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
plt.rcParams.update({'font.size': 12})
```

```python
data = pd.read_csv('../../data/Benchmark/toyproblem.csv',header=None)
```

```python
#PDISAS version
f, ax = plt.subplots(2,1,figsize=[10,4])
ax[0].plot(data.iloc[:,0],'k',linewidth=0.5)
ax[0].set_xlim(ax[0].get_xlim())
ax[0].set_ylim(ax[0].get_ylim())
var1_axis = ax[0].get_ylim()
ax[0].plot([ax[0].get_xlim()[0],0],[5,5],color='grey',lw=0.5,alpha=0.5)
ax[0].plot([ax[0].get_xlim()[0],300],[10,10],color='grey',lw=0.5,alpha=0.5)
ax[0].plot([ax[0].get_xlim()[0],600],[15,15],color='grey',lw=0.5,alpha=0.5)
ax[0].set_ylabel(r'$x_1$')
ax[0].set_yticks([5,10,15,20])

ax[1].plot(data.iloc[:,1],'k',linewidth=0.5)
ax[1].set_xlim(ax[1].get_xlim())
ax[1].set_ylim(ax[1].get_ylim())
var2_axis = ax[1].get_ylim()
ax[1].plot([ax[1].get_xlim()[0],0],[2,2],color='grey',lw=0.5,alpha=0.5)
ax[1].plot([ax[1].get_xlim()[0],300],[3,3],color='grey',lw=0.5,alpha=0.5)
ax[1].plot([ax[1].get_xlim()[0],600],[4,4],color='grey',lw=0.5,alpha=0.5)
ax[1].set_xlabel('data index')
ax[1].set_ylabel(r'$x_2$')
ax[1].set_yticks([1,2,3,4,5])

#ax[0].fill(np.concatenate([[0,300], [300,0]]),
#                 np.concatenate([[ax[0].get_ylim()[0],ax[0].get_ylim()[0]],[ax[0].get_ylim()[1],ax[0].get_ylim()[1]]]),
#                 alpha=.25, fc='grey', ec='None')
#ax[0].fill(np.concatenate([[600,1000], [1000,600]]),
#                 np.concatenate([[ax[0].get_ylim()[0],ax[0].get_ylim()[0]],[ax[0].get_ylim()[1],ax[0].get_ylim()[1]]]),
#                 alpha=.25, fc='grey', ec='None')
#ax[1].fill(np.concatenate([[300,600], [600,300]]),
#                 np.concatenate([[ax[1].get_ylim()[0],ax[1].get_ylim()[0]],[ax[1].get_ylim()[1],ax[1].get_ylim()[1]]]),
#                 alpha=.25, fc='grey', ec='None')
plt.tight_layout()
plt.savefig('../../JMSE_figures/bc.png',dpi=300)
plt.show()
```

```python
#Slide version
plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(1,1,figsize=[6,2])
x = np.linspace(st.gamma.ppf(0.1,10,scale=0.5),st.gamma.ppf(0.9,10,scale=0.5),100)
ax.plot(x, st.gamma.pdf(x,10,scale=0.5),'k')
ax.set_xlim([st.gamma.ppf(0.1,10,scale=0.5),st.gamma.ppf(0.9,10,scale=0.5)])
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks([])
plt.tight_layout()
plt.savefig('../../plots/slide/gamma1.png',dpi=300)
plt.show()

plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(1,1,figsize=[2,6])
x = np.linspace(st.lognorm.ppf(0.1,0.2462,scale=np.exp(0.6628)),st.lognorm.ppf(0.9,0.2462,scale=np.exp(0.6628)),100)
ax.plot(st.lognorm.pdf(x,0.2462,scale=np.exp(0.6628)),x,'k')
ax.set_ylim([st.lognorm.ppf(0.1,0.2462,scale=np.exp(0.6628)),st.lognorm.ppf(0.9,0.2462,scale=np.exp(0.6628))])
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.yaxis.set_ticks([])
ax.set_xticklabels(ax.get_xticks(),rotation=270)
plt.tight_layout()
plt.savefig('../../plots/slide/lognorm1.png',dpi=300)
plt.show()

plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(1,1,figsize=[6,2])
x = np.linspace(st.gamma.ppf(0.1,40,scale=0.25),st.gamma.ppf(0.9,40,scale=0.25),100)
ax.plot(x, st.gamma.pdf(x,40,scale=0.25),'k')
ax.set_xlim([st.gamma.ppf(0.1,40,scale=0.25),st.gamma.ppf(0.9,40,scale=0.25)])
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks([])
plt.tight_layout()
plt.savefig('../../plots/slide/gamma2.png',dpi=300)
plt.show()

plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(1,1,figsize=[2,6])
x = np.linspace(st.lognorm.ppf(0.1,0.1655,scale=np.exp(1.0849)),st.lognorm.ppf(0.9,0.1655,scale=np.exp(1.0849)),100)
ax.plot(st.lognorm.pdf(x,0.1655,scale=np.exp(1.0849)),x,'k')
ax.set_ylim([st.lognorm.ppf(0.1,0.1655,scale=np.exp(1.0849)),st.lognorm.ppf(0.9,0.1655,scale=np.exp(1.0849))])
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.yaxis.set_ticks([])
ax.set_xticklabels(ax.get_xticks(),rotation=270)
plt.tight_layout()
plt.savefig('../../plots/slide/lognorm2.png',dpi=300)
plt.show()

plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(1,1,figsize=[6,2])
x = np.linspace(st.gamma.ppf(0.1,100,scale=0.15),st.gamma.ppf(0.9,100,scale=0.15),100)
ax.plot(x, st.gamma.pdf(x,100,scale=0.15),'k')
ax.set_xlim([st.gamma.ppf(0.1,100,scale=0.15),st.gamma.ppf(0.9,100,scale=0.15)])
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks([])
plt.tight_layout()
plt.savefig('../../plots/slide/gamma3.png',dpi=300)
plt.show()

plt.rcParams.update({'font.size': 16})
fig, ax = plt.subplots(1,1,figsize=[2,6])
x = np.linspace(st.lognorm.ppf(0.1,0.1245,scale=np.exp(1.3785)),st.lognorm.ppf(0.9,0.1245,scale=np.exp(1.3785)),100)
ax.plot(st.lognorm.pdf(x,0.1245,scale=np.exp(1.3785)),x,'k')
ax.set_ylim([st.lognorm.ppf(0.1,0.1245,scale=np.exp(1.3785)),st.lognorm.ppf(0.9,0.1245,scale=np.exp(1.3785))])
# Hide the right and top spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.yaxis.set_ticks([])
ax.set_xticklabels(ax.get_xticks(),rotation=270)
plt.tight_layout()
plt.savefig('../../plots/slide/lognorm3.png',dpi=300)
plt.show()
```

```python
f, ax = plt.subplots(2,2,figsize=[10,4])
ax[0,0].plot(range(500),data.iloc[:500,0],'k',linewidth=0.5)
ax[0,0].set_ylabel(r'$x_1$')
ax[0,0].set_ylim(var1_axis)
ax[0,0].axes.xaxis.set_ticklabels([])
ax[0,0].set_yticks([3,6,9,12])

ax[0,1].plot(range(500,1000),data.iloc[500:,0],'k',linewidth=0.5)
ax[0,1].set_ylim(var1_axis)
ax[0,1].axes.yaxis.set_ticklabels([])
ax[0,1].axes.xaxis.set_ticklabels([])

ax[1,0].plot(range(500),data.iloc[:500,1],'k',linewidth=0.5)
ax[1,0].set_xlabel(r'$X_1$')
ax[1,0].set_ylabel(r'$x_2$')
ax[1,0].set_ylim(var2_axis)
ax[1,0].set_yticks([1,2,3,4,5])

ax[1,1].plot(range(500,1000),data.iloc[500:,1],'k',linewidth=0.5)
ax[1,1].set_xlabel(r'$X_2$')
ax[1,1].set_ylim(var2_axis)
ax[1,1].axes.yaxis.set_ticklabels([])

plt.tight_layout()
#plt.savefig('../plots/bc_k.png',dpi=300)
plt.show()
```
